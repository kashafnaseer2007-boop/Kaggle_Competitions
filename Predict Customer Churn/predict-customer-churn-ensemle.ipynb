```python
import pandas as pd
import numpy as np
```

```python
# Load original test.csv to get the IDs (order must match)
test_original = pd.read_csv('/kaggle/input/competitions/playground-series-s6e3/test.csv')
test_ids = test_original['id']

# Load your previous submissions
xgb_sub = pd.read_csv('/kaggle/input/datasets/kashafrana2007/my-xgb-lgb-submissions/xgb_submission.csv') 
lgb_sub = pd.read_csv('/kaggle/input/datasets/kashafrana2007/my-xgb-lgb-submissions/lgb_submission.csv') 

# Verify they have the same IDs and order
assert (xgb_sub['id'] == test_ids).all(), "XGBoost ID mismatch!"
assert (lgb_sub['id'] == test_ids).all(), "LightGBM ID mismatch!"
```

```python
ensemble_preds = (xgb_sub['Churn'] + lgb_sub['Churn']) / 2

ensemble_sub = pd.DataFrame({
    'id': test_ids,
    'Churn': ensemble_preds
})
ensemble_sub.to_csv('ensemble_simple_avg.csv', index=False)
print("Simple average ensemble saved.")
```

```python
# Replace with your actual CV scores from each notebook
xgb_cv = 0.91363   # your best XGBoost CV AUC
lgb_cv = 0.91378   # your best LightGBM CV AUC 

total = xgb_cv + lgb_cv
w_xgb = xgb_cv / total
w_lgb = lgb_cv / total

print(f"Weights: XGBoost = {w_xgb:.4f}, LightGBM = {w_lgb:.4f}")

weighted_preds = w_xgb * xgb_sub['Churn'] + w_lgb * lgb_sub['Churn']

weighted_sub = pd.DataFrame({
    'id': test_ids,
    'Churn': weighted_preds
})
weighted_sub.to_csv('ensemble_weighted.csv', index=False)
print("Weighted ensemble saved.")
```
